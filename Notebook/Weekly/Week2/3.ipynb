```python
#3
def separate(student, group_size):
    if student % group_size == 1:
        return f"There will be {student // group_size} Groups with {student % group_size} student Left over."
    else:
        return f"There will be {student // group_size} Groups with {student % group_size} students Left over."
    
student = int(input("Enter number of students: "))
group_size = int(input("Enter group size: "))
print(separate(student, group_size))
```
