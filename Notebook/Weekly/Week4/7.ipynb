```python
#7
import statistics 

temperatures = []

for i in range(6):
    temp = float(input(f'Enter temperature {i + 1} in Celsius: '))
    temperatures.append(temp)

maximun = max(temperatures)
minimun = min(temperatures)
mean = statistics.mean(temperatures)

print(f'Maximum temperature: {maximun}C')
print(f'Minimum temperature: {minimun}C')
print(f'Mean temperature: {mean}C')
```
