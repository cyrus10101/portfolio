```python
#6
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

temp_input = input("Enter temperature in Celsius (e.g: 25c): ").strip()
if temp_input[-1].lower() == 'c':
    try:
        celsius_value = float(temp_input[:-1])
        fahrenheit_value = celsius_to_fahrenheit(celsius_value)
        print(f"{celsius_value}C is equal to {fahrenheit_value}F")
    except ValueError:
        print("Invalid temperature value. Please enter a valid number followed by 'c'.")
else:
    print("Invalid format. Please ensure the temperature ends with 'c'.")       
```
