```python
'''
6. Write a program that decrypts messages encoded as above.
'''

def decrypt_message(encrypted, interval):
    decrypted = ""
    for i in range(0, len(encrypted), interval):
        if i < len(encrypted):
            decrypted += encrypted[i]
    return decrypted

encrypted = "sxyexynxydxy cxyhxyexyexysxye"
interval = 2
result = decrypt_message(encrypted, interval)
print(f"Decrypted: {result}")
```
