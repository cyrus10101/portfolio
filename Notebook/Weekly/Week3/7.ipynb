```python
#7
while True:
    num = int(input("Enter a number to see its multiplication table from (0-12): "))

    if num < 0 or num > 12:
        print("Please enter a valid number between 0 and 12.")
        
    else:
        for i in range(0, 13):
            print(f"{i} * {num} = {i * num}")
        break
```
