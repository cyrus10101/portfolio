```python
#3
def check_password(pw, cpw):
    return pw == cpw

password = input("Passwowrd(case sensetive): ")
c_password = input("Confirm Passwowrd(case sensetive): ")

if 8 <= len(password) <= 12 and 8 <= len(c_password) <= 12:
    print("Valid Password Length")
else:
    print("Error in Password Length")
    
if check_password(password, c_password):
    print("Password Set")
else:
    print("Error! Password not matched")
```
