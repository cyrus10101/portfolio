```python
#5
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

def check_password(pw, cpw):
    return pw == cpw

while True:
    password = input("Password (case sensitive): ")
    c_password = input("Confirm Password (case sensitive): ")

    if password.strip().lower() in [bad.lower() for bad in BAD_PASSWORDS]:
        print("Error! Password is too common.")
        continue

    if not 8 <= len(password) <= 12:
        print("Invalid Password Length. Password must be between 8 to 12 characters.")
        continue

    if not check_password(password, c_password):
        print("Error! Passwords do not match.")
        continue

    print("Password Set Successfully!")
    break
```
