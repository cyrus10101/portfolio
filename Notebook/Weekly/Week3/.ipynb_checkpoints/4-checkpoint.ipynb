```python
#4
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

def check_password(pw, cpw):
    return pw == cpw

password = input("Passwowrd(case sensetive): ")
c_password = input("Confirm Passwowrd(case sensetive): ")

for bad in BAD_PASSWORDS:
    if password.strip().lower() == bad.strip().lower():
        print("Error! Password is too common.")
        continue
    else:   
        print("Password Accepted")
        break

if 8 <= len(password) <= 12 and 8 <= len(c_password) <= 12:
    print("Valid Password Length. pasword must be between 8 to 12 characters")
    
if not check_password(password, c_password):
    print("Error! Password not matched")


```
