```python
#8
while True:
    number = int(input(
        "Enter a number from 0-12\n"
        "Enter a negative number to print the table backwards\n"
        "Enter 99 to exit: "
    ))

    if number == 99:
        print("Exiting the program.")
        break

    if number >= 0:
        print(f"\nMultiplication table for {number} (forward):")
        for i in range(0, 13):
            print(f"{i} * {number} = {i * number}")
    else:
        print(f"\nMultiplication table for {abs(number)} (backward):")
        for i in range(12, -1, -1):
            print(f"{i} * {abs(number)} = {i * abs(number)}")
    print("\n" + "-"*30 + "\n")
```
