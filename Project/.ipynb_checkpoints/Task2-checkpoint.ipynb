```python
# Introduction to Programming
# Assignment: Task 2
# The Problem
# In this task you should create a program that asks a user to enter their password, checks its length, and then verifies it further by asking for various characters from it.
# It should work like this. First, the user should be prompted to enter their password (do not worry that it appears on the screen). Then the program should check the length of the password, and if it is less than 9 characters long (so a length of 9 is OK, 8 is not OK), the user should be informed, and the program should exit.
# If the password is sufficiently long, the program should ask for three random letters from the password (as shown below). If any of these are given incorrectly, the program should exit immediately. (Do not worry if the program asks for the same letter more than once.)

password = input("Enter your password: ")

password_length = len(password)

if password_length < 9:
    print()
    print("Password too short.")
else:
    pos1 = password_length // 3
    if pos1 == 0:
        pos1 = 1
    pos2 = password_length // 2
    if pos2 == 0:
        pos2 = 1
    pos3 = password_length - 1
    if pos3 < 1:
        pos3 = 1
    
    letter1 = password[pos1 - 1]
    letter2 = password[pos2 - 1]
    letter3 = password[pos3 - 1]
    
    answer1 = input("Enter letter at position " + str(pos1) + ": ")
    print()
    if answer1 == letter1:
        print("Correct")
        print()
        answer2 = input("Enter letter at position " + str(pos2) + ": ")
        print()
        if answer2 == letter2:
            print("Correct")
            print()
            answer3 = input("Enter letter at position " + str(pos3) + ": ")
            print()
            if answer3 == letter3:
                print("Correct")
                print()
                print("Security check passed.")
            else:
                print("Security check failed.")
        else:
            print("Security check failed.")
    else:
        print("Security check failed.")
```
