```python
# Introduction to Programming
# Assignment: Task 1
# The Problem
# Pizza takeaways are popular. There is a wide variety from which to choose, ranging from international franchises to small local businesses.
# For reasons that are not clear, the pricing of takeaway pizzas is very rarely straightforward. Prices vary depending on many factors. These might include the number ordered, the time of day the order is placed, the day of the week itself, whether the order will be collected, or even whether an app was used to place the order.
# A computer program is often the only practical way to work out the price of a takeaway pizza order.
# The Task
# In this task you will develop a program that will assist one Pizza Provider to calculate how much to charge a customer for a given order. The rules to be applied are set out below.
# Beckett Pizza Plaza (henceforth BPP) offers a range of pizzas for collection or delivery. In common with all others in the marketplace, the pricing of individual pizzas is simple, but things become more complex with a complete order. In an attempt to simplify this needlessly complex state of affairs, BPP has decided to offer only one discount.
# The customer must order FOUR pizzas. The pizzas are priced differently, but the cheapest will not be charged for.
# This would commonly be called a "Four-for-Three" offer.
# Your task is to create a program that accepts four pizza prices, and carries out the required calculation to determine what the customer should be charged. The program should also display the discount the customer has achieved (which is at most 25%).

print("Beckett Pizza Plaza 4-for-3 Offer")
print("=================================")
print()

prices = []
pizza_number = 1

while pizza_number <= 4:
    user_input = input("Enter Price of Pizza #" + str(pizza_number) + ": ")
    valid = True
    price_value = 0
    
    try:
        price_value = float(user_input)
        if price_value <= 0:
            valid = False
    except:
        valid = False
    
    if valid:
        prices.append(price_value)
        pizza_number = pizza_number + 1
    else:
        print("Please enter a valid price!")

cheapest = prices[0]
if prices[1] < cheapest:
    cheapest = prices[1]
if prices[2] < cheapest:
    cheapest = prices[2]
if prices[3] < cheapest:
    cheapest = prices[3]

total = prices[0] + prices[1] + prices[2] + prices[3]
final_total = total - cheapest
original_total = total
discount_amount = cheapest
discount_percent = (discount_amount / original_total) * 100

if discount_percent - int(discount_percent) >= 0.5:
    discount_rounded = int(discount_percent) + 1
else:
    discount_rounded = int(discount_percent)

final_total_str = str(final_total)
if "." in final_total_str:
    parts = final_total_str.split(".")
    if len(parts[1]) == 1:
        final_total_str = final_total_str + "0"
    elif len(parts[1]) > 2:
        rounded = round(final_total, 2)
        final_total_str = str(rounded)
        if "." in final_total_str:
            parts = final_total_str.split(".")
            if len(parts[1]) == 1:
                final_total_str = final_total_str + "0"
else:
    final_total_str = final_total_str + ".00"

print()
print("Order Total is £" + final_total_str + ", a fabulous discount of " + str(discount_rounded) + "%!")
```
